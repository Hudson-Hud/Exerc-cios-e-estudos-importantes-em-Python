## 3\. Strings

```
# Aplicando 3 métodos de string distintos

cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'
testeCount = cancao.count('roda')
print(testeCount)

testeCapitalize = cancao.capitalize()
print(testeCapitalize)

testeFimDaFrase = cancao.endswith('.')
print(testeFimDaFrase)
```

Extraia da string abaixo o valor da taxa **selic** na variável `selic` e o valor do **ano** na variavel `ano`. Imprima os valores na tela.

```
noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'

# Comando pra achar o início do fatiamento da variável selic
achar_taxa = noticia.find('2,75%')
print(achar_taxa)

selic = noticia[12:17]
print(selic)

# Comando pra achar o início do fatiamento da variável ano
achar_ano = noticia.find('6')
print(achar_ano)

ano = noticia[62:-1]
print(ano)
```
